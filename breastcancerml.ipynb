```python
X=data_frame.drop(columns='label',axis=1)
Y=data_frame['label']
```

```python
from sklearn.feature_selection import SelectKBest, f_classif
import pandas as pd

# X = your input feaures before scaling
# y = your target (diagnosis)
selector = SelectKBest(score_func=f_classif, k=10)
selector.fit(X, Y)

top_features = X.columns[selector.get_support()]
print("Top 10 features:", list(top_features))
```

```python
top_10_features = [
    'mean radius', 'mean perimeter', 'mean area', 'mean concavity', 'mean concave points',
    'worst radius', 'worst perimeter', 'worst area', 'worst concavity', 'worst concave points'
]

X_selected = X[top_10_features]
```

Scaling the features of the data

```python
scaler=StandardScaler()
X_scaled=scaler.fit_transform(X_selected)
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_scaled,Y,test_size=0.3,random_state=42)
```

**Model training**

1. Logistic Regression

```python
model1=LogisticRegression(max_iter=1000)
model1.fit(X_train,y_train)
```

```python
y_pred_lr=model1.predict(X_test)
testing_data_accuracy1=accuracy_score(y_test,y_pred_lr)
testing_data_cm=confusion_matrix(y_test,y_pred_lr)
testing_data_cr=classification_report(y_test,y_pred_lr)
```

```python
print(testing_data_accuracy1)
print(testing_data_cm)
print(testing_data_cr)
```

2. KNN - K Neighbors Classifier

```python
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,y_train)
```

```python
#prediction and evaluation of knn model
y_pred_knn=knn.predict(X_test)
knn_accuracy=accuracy_score(y_test,y_pred_knn)
confusion_matrix_knn=confusion_matrix(y_test,y_pred_knn)
classification_report_knn=classification_report(y_test,y_pred_knn)
```

```python
print(knn_accuracy)
print(confusion_matrix_knn)
print(classification_report_knn)
```

3. Random Forest

```python
rf_model=RandomForestClassifier(n_estimators=100,random_state=2)
rf_model.fit(X_train,y_train)
```

```python
#prediction and evaluation of random forest model
y_prediction_rf=rf_model.predict(X_test)
acc_rf=accuracy_score(y_test,y_prediction_rf)
confusion_matrix_rf=confusion_matrix(y_test,y_prediction_rf)
classification_report_rf=classification_report(y_test,y_prediction_rf)
```

```python
print(acc_rf)
print(confusion_matrix_rf)
print(classification_report_rf)
```

4. SVM - support vector machine

```python
svm=SVC(kernel='linear',probability=True)
svm.fit(X_train,y_train)
```

```python
y_prediction_svm=svm.predict(X_test)
svm_acc=accuracy_score(y_test,y_prediction_svm)
confusion_matrix_svm=confusion_matrix(y_test,y_prediction_svm)
classification_report_svm=classification_report(y_test,y_prediction_svm)
```

```python
print(svm_acc)
print(confusion_matrix_svm)
print(classification_report_svm)
```

Build a predictive system

**Grouped Bar Chart for visually comparing performance of the above Tested Models**

```python
model_metrics={ 
    "Logistic Regression": {"accuracy": 0.9474, "precision": 0.95, "recall": 0.95, "f1": 0.95},
    "KNN": {"accuracy": 0.9824, "precision": 0.98, "recall": 0.98, "f1": 0.98},
    "Random Forest": {"accuracy": 0.9591, "precision": 0.96, "recall": 0.96, "f1": 0.96},
    "SVM": {"accuracy": 0.9708, "precision": 0.97, "recall": 0.97, "f1": 0.97}
}
# Models and metrics
models = list(model_metrics.keys())
metrics = ["accuracy", "precision", "recall", "f1"]
n_models = len(models)
bar_width = 0.2
x = np.arange(n_models)

# Set up bar positions for each metric group
offsets = [-1.5, -0.5, 0.5, 1.5]
colors = ['skyblue', 'lightgreen', 'salmon', 'mediumpurple']

# Plot grouped bars
plt.figure(figsize=(10, 6))
for i, metric in enumerate(metrics):
    scores = [model_metrics[model][metric] for model in models]
    plt.bar(x + offsets[i]*bar_width, scores, width=bar_width, label=metric.capitalize(), color=colors[i])

# Labeling
plt.xticks(x, models)
plt.ylim(0.85, 1.01)
plt.ylabel("Score")
plt.title("Model Performance Comparison")
plt.legend(title="Metric")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

Plotting the ROC curve

```python
from sklearn.metrics import roc_curve, roc_auc_score

#get probabilities
logistic_probs=model1.predict_proba(X_test)[:,1]
knn_probs=knn.predict_proba(X_test)[:,1]
rf_probs=rf_model.predict_proba(X_test)[:,1]
svm_probs=svm.predict_proba(X_test)[:,1]

#Calculate ROC values
fpr_lr,tpr_lr, _=roc_curve(y_test,logistic_probs)
fpr_knn,tpr_knn, _=roc_curve(y_test,knn_probs)
fpr_rf,tpr_rf, _=roc_curve(y_test,rf_probs)
fpr_svm,tpr_svm, _=roc_curve(y_test,svm_probs)

#Plot the ROC curve
plt.figure(figsize=(10,7))
plt.plot(fpr_lr,tpr_lr, label=f"Logistic Regression (AUC ={roc_auc_score(y_test,logistic_probs):.2f})")
plt.plot(fpr_knn,tpr_knn, label=f"K Neighbors Classifier (AUC ={roc_auc_score(y_test,knn_probs):.2f})")
plt.plot(fpr_rf,tpr_rf, label=f"Random Forest (AUC ={roc_auc_score(y_test,rf_probs):.2f})")
plt.plot(fpr_svm,tpr_svm, label=f"SVM - Support Vector Machine (AUC ={roc_auc_score(y_test,svm_probs):.2f})")

plt.plot([0, 1], [0, 1], 'k--')

plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve Comparison')
plt.legend()
plt.grid()
plt.show()
```
